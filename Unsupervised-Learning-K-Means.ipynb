## Unsupervised Learning 
### Implementation of K-Means Clustering

```python
# pseudocode:
# initialize K-means with random values
# iterate through items 
### find the mean closest to the item by calculating Eucledian distance of item with each of the means
### assign item (data point) to the mean
### update the mean by shifting it the average of the items (data points) in that cluster
```

```python
# importing libraries
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
```

```python
# create the dataset with make_blobs an plot
X, y = make_blobs(n_samples = 500, n_features = 2, centers = 3, random_state = 23)
fig = plt.figure(0)
plt.grid(True)
plt.scatter(X[:,0], X[:,1])
plt.show
```

```python
# initiatialize the random conroids
k = 3
clusters = {}
np.random.seed(23)

for idx in range(k):
    center = 2*(2*np.random.random((X.shape[1],))-1)
    points = []
    cluster = {
        'center': center,
        'points': []
    }
    clusters[idx] = cluster
```

```python
# define Eucledian Distance
def distance(p1, p2):
    return np.sqrt(np.sum((p1-p2)**2))
```

```python
# update cluster centers based on the mean of assigned points
def update_clusters(X, clusters):
    for i in range(k):
        points = np.array(clusters[i]['points'])
        new_center = points.mean(axis=0)
        clusters[i]['center'] = new_center
        clusters[i]['points'] = []

```

```python
# predicting the clusrter for data points
def pred_cluster(X, clusters):
    pred = []
    for i in range(X.shape[0]):
        dist = []
        for j in range(k):
            dist.append(distance(X[i],clusters[j]['center']))
        pred.append(np.argmin(dist))
    return pred 
```

```python
# assign, updating, predicitng the centroids:
assigned_clusters = assign_clusters(X, clusters)
updated_clusters = update_clusters(X, assigned_clusters)
pred = pred_cluster(X, updated_clusters)
```

```python
# plotting the data points with their predicted cluster center
plt.scatter(X[:,0],X[:,1], c = pred)
for i in clusters:
    center = clusters[i]['center']
    plt.scatter(center[0],center[1],marker="^",c='red')
    plt.show()
```
